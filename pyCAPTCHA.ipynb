```python
from captcha.image import ImageCaptcha
from captcha.audio import AudioCaptcha

from IPython.display import Image
from IPython.display import Audio

import ipywidgets as widgets
from ipywidgets import interact, interact_manual

import string
import random
```

```python
image = ImageCaptcha(width = 300, height = 110)
audio = AudioCaptcha()
```

```python
captcha_string_alphanumeric = ''.join(random.SystemRandom().choice(string.ascii_uppercase + string.ascii_lowercase + string.digits) for _ in range(7))
# '9w3rTy'
captcha_string_numeric = ''.join(random.SystemRandom().choice(string.digits) for _ in range(3))
# '189'
```

```python
image.write(captcha_string_alphanumeric, 'captcha.png')
audio.write(captcha_string_numeric, 'captcha.wav')
```

```python
display(Image('captcha.png'))
print()

image_captcha_response = widgets.Text(placeholder= "Captcha Code")
display(image_captcha_response)

def callback(wdgt):
    captcha_response_alphanumeric = wdgt.value
    if captcha_response_alphanumeric == captcha_string_alphanumeric:
        print("✔️ verified: '" + wdgt.value + "' is correct")
    else:
        print("❌ please try again: '" + wdgt.value + "' is incorrect")

image_captcha_response.on_submit(callback)
```

```python
display(Audio('captcha.wav'))
print()

audio_captcha_response = widgets.Text(placeholder= "Captcha Code")
display(audio_captcha_response)

def callback(wdgt):
    captcha_response_numeric = wdgt.value
    if captcha_response_numeric == captcha_string_numeric:
        print("✔️ verified: '" + wdgt.value + "' is correct")
    else:
        print("❌ please try again: '" + wdgt.value + "' is incorrect")

audio_captcha_response.on_submit(callback)
```
